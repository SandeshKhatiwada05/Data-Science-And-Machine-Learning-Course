  
#### Sample size should not be more than 10% of population.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_size = [np.random.randint(1, 1000) for i in range (10000)]
pop_size
```

```python
pop_table = pd.DataFrame({"pop_data": pop_size})
pop_table
```

```python
plt.figure(figsize=(4,4))
sns.kdeplot(x="pop_data", data=pop_table)
```
