  
#### Sample size should not be more than 10% of population.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_size = [np.random.randint(1, 1000) for i in range (10000)]
pop_size
```

```python
pop_table = pd.DataFrame({"pop_data": pop_size})
pop_table
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="pop_data", data=pop_table)
```

### Taking sample mean from population to check, we take 50 samples with 500 data in each 

```python
sample_mean= []
for no_sample in range (60):
    sample_data = []
    for sampledata_inside_sample in range (500):
        sample_data.append(np.random.choice(pop_size))
    sample_mean.append(np.mean((sample_data)))
sample_mean
```

```python
sample_mean_for_plotting = pd.DataFrame({"sample_mean_comined": sample_mean})
sample_mean_for_plotting
```

```python
plt.figure(figsize=(2,2))
sns.kdeplot(x="sample_mean_comined", data=sample_mean_for_plotting)
```
