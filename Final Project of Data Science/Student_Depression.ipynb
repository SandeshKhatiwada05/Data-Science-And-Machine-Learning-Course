## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le_gender = LabelEncoder()
le_city = LabelEncoder()
le_profession = LabelEncoder()
le_sleepDuty = LabelEncoder()
le_Diet = LabelEncoder()
le_Degee = LabelEncoder()
le_sucidal_thoughts = LabelEncoder()
le_family_mental_illness = LabelEncoder()
```

```python
dataset["Gender"] = le_gender.fit_transform(dataset["Gender"])
dataset["City"] = le_city.fit_transform(dataset["City"])
dataset["Profession"] = le_profession.fit_transform(dataset["Profession"])
dataset["Sleep Duration"] = le_sleepDuty.fit_transform(dataset["Sleep Duration"])
dataset["Dietary Habits"] = le_Diet.fit_transform(dataset["Dietary Habits"])
dataset["Degree"] = le_Degee.fit_transform(dataset["Degree"])
dataset["Have you ever had suicidal thoughts ?"] = le_sucidal_thoughts.fit_transform(dataset["Have you ever had suicidal thoughts ?"])
dataset["Family History of Mental Illness"] = le_family_mental_illness.fit_transform(dataset["Family History of Mental Illness"])
```

```python
import joblib

# Save all your encoders in a dictionary
encoders = {
    'Gender': le_gender,
    'City': le_city,
    'Profession': le_profession,
    'Sleep Duration': le_sleepDuty,
    'Dietary Habits':  le_Diet,
    'Degree': le_Degee,
    'Have you ever had suicidal thoughts ? ': le_sucidal_thoughts,
    'Family History of Mental Illness': le_family_mental_illness
}

joblib.dump(encoders, 'encoders.pkl')
print("Encoders saved")
```

### Algorithm

```python
X = dataset.drop(columns=['id', 'Depression'])
y = dataset['Depression']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# help(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
import joblib
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Create pipeline that does scaling + model prediction
pipeline = Pipeline([
    ('scaler', StandardScaler()),  
    ('model', lr)
])

# Fit pipeline on training data
pipeline.fit(X_train, y_train)

# Save entire pipeline
joblib.dump(pipeline, 'depression_pipeline. pkl')

print("Pipeline saved successfully")
```
