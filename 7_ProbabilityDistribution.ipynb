#### 2 Dice Probability Distribution


#### 1. A Probability Density Function (PDF) describes the likelihood of a continuous random variable falling within a given range of values, with the area under the curve representing the probability.


#### 2. A probability mass function (PMF) gives the probability that a discrete random variable is exactly equal to some value, often denoted as \(P(X=x)\). A valid PMF must have probabilities between 0 and 1, inclusive, and the sum of all probabilities for every possible value must equal 1.
