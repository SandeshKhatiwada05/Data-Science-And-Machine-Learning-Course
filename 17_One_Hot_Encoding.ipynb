```python
# We encode columns gender and married, so we fill null values
# Gender 
dataset["Gender"].fillna(dataset["Gender"].mode()[0], inplace=True)

# Married
dataset["Married"].fillna(dataset["Married"].mode()[0], inplace=True)

print("Missing Value filled in categorical data using mode")
```

```python
#to check if gender and married still have null values
dataset.isna().sum()
```
