```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
alldataset = pd.read_csv(r"D:\Github\Data-Science-And-Machine-Learning-Course\Datasets\Social_Network_Ads.csv")
```

```python
alldataset.head(4)
```

```python
dataset = alldataset.drop(columns=["User ID", "Gender", "EstimatedSalary"])
dataset
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(dataset, x= dataset["Age"], y=dataset["Purchased"])
plt.show()
```

```python
X = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# help(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```
