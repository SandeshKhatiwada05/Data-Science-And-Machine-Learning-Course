### Skewness

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv(r"D:\Github\Data-Science-And-Machine-Learning-Course\DSML Class Orchid\Day 4\titanic_train.csv")
```

```python
dataset["Age"].fillna(dataset["Age"].mean(), inplace= True)
```

```python
print("Null in Age: ", dataset["Age"].isna().sum())
```

#### Finding mean, median and mode

```python
mean_age = np.mean(dataset["Age"])
median_age = np.median(dataset["Age"])
mode_age = dataset["Age"].mode()[0]
```

```python
print("Mean: ", mean_age)
print("Median: ", median_age)
print("Mode: ", mode_age)
```

### Finding skewness

```python
if mean_age<median_age and median_age<mode_age and mean_age<mode_age:
    print("It is negatively skewed")
if mean_age==mode_age==median_age:
    print("It is not skewed")
if mode_age<mean_age and mode_age<median_age and median_age<mean_age:
    print("It is postively skewed")
```
