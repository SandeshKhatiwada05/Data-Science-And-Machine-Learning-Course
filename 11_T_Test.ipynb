### For T test:
#### 1. Population mean needed
#### 2. Sample mean needed
#### 3. Sample standard deviation needed
#### 4. sample size < 30 needed


```python
pop_mean = 150

sample_mean = 148
sample_sd = 5

sample_size = 25

alpha = 0.05
```

```python
import scipy.stats as sc
import math
```

```python
t_tabulated = sc.t.ppf(alpha, df=sample_size-1) #alpha, degree of freedom
t_tabulated
```

```python
t_calculated = (sample_mean - pop_mean)/(sample_sd/math.sqrt(sample_size))
t_calculated
```

```python
if t_calculated >  t_tabulated:
    print("Alternative hypothesis (H1) is accepted")
else:
    print("Null hypothesis (H0) is accepted")
```
