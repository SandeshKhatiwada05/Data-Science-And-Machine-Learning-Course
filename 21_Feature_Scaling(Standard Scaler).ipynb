### Standardization  


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\Github\Data-Science-And-Machine-Learning-Course\Datasets\Loan.csv")
```

```python
dataset.head(2)
```

```python
#doing for LoanAmount
dataset["LoanAmount"].isna().sum()
```

```python
dataset["LoanAmount"].fillna(dataset["LoanAmount"].mean(), inplace=True)
```

```python
#checking again
dataset["LoanAmount"].isna().sum()
```

```python
# now scaling using standard scaler
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
ss.fit(dataset[["LoanAmount"]])
```

```python
standard_value = ss.transform(dataset[["LoanAmount"]])
standard_value
```

```python
dataset["Loan Amount Standard"] = standard_value
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["LoanAmount"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Loan Amount Standard"])
plt.show()
```
