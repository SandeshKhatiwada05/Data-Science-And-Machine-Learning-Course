```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"D:\Github\Data-Science-And-Machine-Learning-Course\Datasets\placement.csv")
```

```python
dataset.head(3)
```

```python
dataset.isna().sum()
```

```python
sns.kdeplot(data=dataset["package"])
```

```python
X = dataset.drop(columns=["placement", "package"])
y = dataset["placement"]
X.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gb = GaussianNB()
```

```python
gb.fit(X_train, y_train)
```

```python
gb.score(X_test, y_test)
```

```python
plot_decision_regions(X.to_numpy(), y.to_numpy(), clf=gb)
plt.show()
```

```python
mb = MultinomialNB()
```

```python
mb.fit(X_train, y_train)
```

```python
mb.score(X_test, y_test)
```

```python
plot_decision_regions(X.to_numpy(), y.to_numpy(), clf=mb)
plt.show()
```

```python
bb = BernoulliNB()
```

```python
bb.fit(X_train, y_train)
```

```python
bb.score(X_test, y_test)
```
