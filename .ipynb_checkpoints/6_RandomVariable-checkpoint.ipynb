##### If sample space is written in a variable then it is discrete random variable
for example: while rolling a dice it appears {1,2,3,4,5,6,} it is sample space  
if represented as dice_value ={{1,2,3,4,5,6}} this is now discrete random variable

```python
import random
```

```python
for i in range(1,10):
    print(random.random())
```

```python
random.randint(1,10) #random int from 1 to 10
```

```python
random.uniform(1,100) #random float between 1 and 100
```
